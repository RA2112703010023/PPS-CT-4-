The above code are used under the topic of pandas to identify different charactertistics of the above table which was found from an government website. The various functions used are :- 
1) max function - in this function we choose a coloumn and the fucntion will output the biggest number in that coloumn.
2) min function - in this function we choose a coloumn and the function will output the smalled number in the coloumn.
3). mean function - in this function we choose the coloumn and the fuction will output the mean of all the numbers in the coloumn.
4) memory usage - this is a function which is used for the user to identify the memory used by the above database.
5) valuecounts function - This function will output the number of values in the given category.
6) describe function - is a function to output the major charactertics of the database.
